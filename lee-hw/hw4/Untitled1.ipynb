```python
def read_data(path):   
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        line = [line.strip('\n') for line in lines]
        line = [line.split(' ') for line in line]
        X = [x[2:] for x in line]
        return X
```
