```python
import pandas as pd
import numpy as np
from itertools import chain
```

```python
df = pd.read_csv("./train.csv", engine='python', encoding="utf-8")
```

```python
df = df[df['observation']!="RAINFALL"]
```

```python
df.drop(['Date', 'stations', 'observation'], axis=1, inplace=True)
df.head()
```

```python
X_train = []
y_train = []
```

```python
ItemNum = 17
for i in range(int(len(df)/ItemNum)):
    temp_data = df[i*ItemNum:(i+1)*ItemNum]
    for j in range(15):
        X = []
        for k in range(17):
            x = temp_data.iloc[k, j:j+9]
            X.append(x)
        X = list(chain(*X))
        X_train.append(X)
        y = temp_data.iloc[9, j+9]
        y_train.append(y)
```

```python
X_train = np.array(X_train, dtype='float')
y_train = np.array(y_train, dtype='float')
print(X_train.shape)
print(y_train.shape)
print(X_train.dtype)
print(y_train.dtype)
```

```python
from playML.LinearRegression import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit_gd(X_train, y_train, eta=0.000001)
print(lin_reg.coef_)
print(lin_reg.score(X_train, y_train))
```
