```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("./train.csv", engine='python', encoding="utf-8")
df.head(20)
```

```python
df = df[df['observation']=='PM2.5']
df = df.drop(['Date', 'stations', 'observation'], axis=1)
```

```python
X_train = []
y_train = []

for i in range(15):
    X = df.iloc[:,i:i+9]
    X_train.append(X)
    y_train.append(df.iloc[:,i+9])
X_train = np.array(X_train, dtype='float').reshape(3600,9)
y_train = np.array(y_train, dtype='float').reshape(3600)
```

```python
from playML.LinearRegression import LinearRegression
from sklearn.preprocessing import StandardScaler

lin_reg = LinearRegression()
lin_reg.fit_gd(X_train, y_train,eta=0.0001)
lin_reg.coef_
```

```python
lin_reg.score(X_train, y_train)
```

```python
df_test = pd.read_csv("./test.csv", engine='python', encoding="utf-8", header=None)
df_test = df_test[df_test[1]=='PM2.5']
df_test.drop([0,1], axis=1, inplace=True)
df_test.head()
```

```python
df_result = pd.read_csv("./result.csv", engine='python', encoding="utf-8")
df_result.drop(['Unnamed: 0','id'], axis=1, inplace=True)
df_result.head()
```

```python
X_test = np.array(df_test, dtype='float')
y_test = np.array(df_result, dtype='float')
```

```python
lin_reg.predict(X_test)
```
